```python
all_data['Month']= all_data['Order Date'].str[0:2]
all_data.head()
```

# Clean up the Data!

```python
#Check the location of Null/Nan
```

```python
null_df = all_data[all_data.isna().any(axis=1)] 
null_df.head()
```

```python
#Drop Columns if Nan/Null 
```

```python
all_data = all_data.dropna(how = "all" )
all_data.head()
```

```python
#Find Or 
```

```python
temp_df = all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head(6)
```

```python
#Delete Or
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
all_data.head()
```

# Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
```

```python
all_data.head()
```

# Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #
all_data['Price Each']=  pd.to_numeric(all_data['Price Each'])

all_data.head()
```

```python
#Add a Sales Column
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] *all_data['Price Each']
all_data.head()
```

# Q1:What was the best month for sales? How much was earned the month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)Million')
plt.xlabel('Month Number')



plt.show()
```

```python
months = range(1,13)
plt.pie(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)Million')
plt.xlabel('Month Number', size=8)
plt.title('Sales in 12 Months ')
autopct='%.2f%%'
plt.show()
```

# Question 3: When should we display advertisements to maximize likelihood of customers buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour

all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

# Question 4: What product are most often solve together?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
